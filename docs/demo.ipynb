```python
from isoc.mist_config import configuration, update_config

# View current defaults
for k, v in configuration['defaults'].items():
    print(f'{k:20s} = {v}')

# View available rotation options
print('\nRotation:', configuration.get('rotation', {}).get('v_div_vcrit', []), '\n')

# View available photometric systems
for label, value in configuration.get('output', {}).get('output', []):
    print(f'  {value:20s} - {label}')
```

```python
i = Isochrone(database="padova")
ip_legacy = i.query_ezpadova(age_yr=(1e9,1e9,0), Z=(0.3, 0.3, 0))
ip_legacy.data
```

```python
i = Isochrone(database="padova")
ip = i.from_padova(age=(1e9, 1e10, 0.5e9), metallicity=0.02, metallicity_type="Z")
ip.data
```

```python
print(ip.columns)
print("\nTable has {ip.n_isochrones} isochrones")

for val in [
 'log_L',
 'log_Teff',
 'log_age',
 'log_g',
 'mass',
 'metallicity',
]:
    print(f'\n{val}[:5]', getattr(ip, val)[:5])
```

```python
print('\nphotometry data', ip.photometry)     
print('\nUmag[:5]', ip.photometry.Umag[:5])
print('\nVmag[:5]', ip.photometry.Vmag[:5])
print('\nU-V color[:5]', ip.photometry.get_color("Umag", "Vmag")[:5])
print('\nnew photometry data', ip.photometry)     
```

```python
grid = ip.to_isochrones_grid(bands=['Umag', 'Vmag'])
grid
```
